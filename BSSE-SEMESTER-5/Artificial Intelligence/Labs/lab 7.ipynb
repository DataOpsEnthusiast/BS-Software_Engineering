```python
import matplotlib.pyplot as plt
```

```python
xaxis = np.linspace(0,10,1000)
yaxis = np.sin(xaxis)
```

```python
plt.plot(xaxis, yaxis, color = 'red')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.scatter(xaxis, yaxis, color = 'red')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.scatter(xaxis, yaxis, color = 'red')
plt.bar(xaxis, yaxis, color = 'blue')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.scatter(xaxis, yaxis, color = 'red')
plt.bar(xaxis, yaxis, color = 'blue')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.legend(['Scatter', 'Bar'])
plt.show()
```
